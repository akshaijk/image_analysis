```python
from tifffile import imwrite
```

```python
def convert_to_8bit(array):
    # Clip the array values to the 8-bit range (0 to 255)
    clipped_array = np.clip(array, 0, 255)
    # Convert the clipped array to 8-bit data type (uint8)
    converted_array = clipped_array.astype(np.uint8)
    return converted_array
```

```python
np_img = convert_to_8bit(np_img)
```

```python
metadata = {'spacing': 0.1 , 'unit': 'um', 'axes': 'ZYX'}
imwrite("/content/drive/MyDrive/multicut_segs", np_img, imagej=True,
        resolution=(1./0.1, 1./0.1), metadata = metadata)
```
